## Filling missing date with mode

```python
date_mode = df['Date'].mode()[0]
df['Date'] = df['Date'].fillna(date_mode)
```

## Filling missing Product Category with mode

```python
cat_mode = df['Product_Category'].mode()[0]
df['Product_Category'] = df['Product_Category'].fillna(cat_mode)
```

## Converting date format from string to datetime

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y', errors='coerce')
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
print(df['Date'].head(5))
print("Null dates:", df['Date'].isna().sum())
```

## Filling missing values of Sales Amount with mean

```python
sales_mean = df['Sales_Amount'].mean()
df['Sales_Amount'] = df['Sales_Amount'].fillna(sales_mean)
```

# 2. Sales Analysis
